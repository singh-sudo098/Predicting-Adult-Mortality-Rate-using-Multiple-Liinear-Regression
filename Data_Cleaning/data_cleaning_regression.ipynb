```python
import pandas as pd
import numpy as np
```

# Dataset-1

## Life expectancy at birth (years)

```python
df = pd.read_csv("Life expectancy and Healthy life expectancy.csv")
print(df.shape)
# print(df.info())
life_expectancy_df = df.loc[(df['Period'] == 2012) 
                               & (df['Indicator'] == "Life expectancy at birth (years)")
                               & (df['Dim1'] == "Both sexes")
                              ]
print(life_expectancy_df.shape)
countries_life_expectancy = life_expectancy_df["Location"].unique() # actually outputs an array
print(countries_life_expectancy.shape)
```

```python
life_expectancy_df.head()
```

```python
life_expectancy_df.drop(['Indicator', 'Dim1'], axis=1, inplace=True)
life_expectancy_df.rename(columns = {'Location':'Country',
                            'Period':'Year',
                            'First Tooltip':'life_expectancy'}, 
                          inplace = True)
life_expectancy_df.reset_index(drop=True, inplace=True)
life_expectancy_df.head()
```

# Dataset-6

## Total Country Population

```python
# taking log in the case of population will be better
df6 = pd.read_csv("country_population.csv")
print(df6.shape)
# print(df6.info())
pop_df = df6[['Country Name', '2012']]
print(pop_df.shape)
countries_pop = pop_df["Country Name"].unique() # actually outputs an array
print(countries_pop.shape)
```

```python
pop_df.head()
```

```python
pop_year = np.zeros(len(pop_df.index), dtype=int)
for i in range(len(pop_df.index)):
    pop_year[i] = 2012

np.shape(pop_year)
```

```python
pop_df = pop_df.assign(Year = pop_year)
pop_df.head()
```

```python
pop_df.rename(columns = {'2012':'total_population',
                        'Country Name': 'Country'}, 
                          inplace = True)
pop_df.reset_index(drop=True, inplace=True)
pop_df = pop_df[['Country', 'Year', 'total_population']]
pop_df.head()
```
